```python
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict

test_data = unpickle('cifar_test_nolabels.pkl')
images = test_data[b'data']
```

```python
import matplotlib.pyplot as plt

def show_image(img, ax):
    reshaped_img = img.reshape((3, 32, 32)).transpose(1, 2, 0)
    ax.imshow(reshaped_img)

nrows = 5
ncols = 5

fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=(nrows, ncols))

multi_offset = 0
beginning = nrows*ncols*multi_offset
for i, ax in enumerate(axs.flat):
    ax.set_axis_off()
    show_image(images[beginning + i], ax)
fig.tight_layout()
```
